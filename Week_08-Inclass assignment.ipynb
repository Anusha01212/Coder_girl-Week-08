# Arithmetic mean

Arithmetic mean is often called as mean or arithmetic average. Arithmetic mean is calculated by adding all the numbers in a given data set and then dividing it by the total number of items within that set.
formula for AM is Mean (xÌ„) = Sum of all observations/ Number of observations

```python
# calculating Arithmetic mean for belize_airbnb_listings data 

df[["price","number_of_reviews"]].mean()
```

# Geometric mean

Geometric mean is the average value or mean which signifies the central tendency of the set of numbers by taking the root of the product of their values. Geometric Mean can be found by multiplying all the numbers in the given data set and take the nth root for the obtained result.

```python
# created column for geometric mean and assigned none to the column
df['Geometric Mean'] = None
df.head(2)
```

```python
# here i have calculated GM for price and minimum_nights coulmns , i gave the coulmn names as index in iloc
df['Geometric Mean'] = stats.gmean(df.iloc[:, 9:11], axis=1)
df.head(2)
```

# Harmonic mean
Harmonic mean is a type of average that is calculated by dividing the number of values in a data series by the sum of the reciprocals (1/x_i) of each value in the data series.

```python
df['Harmonic Mean'] = stats.hmean(df.iloc[:, 9:11], axis=1)
df.head(2)
```
